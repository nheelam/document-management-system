```python
%%writefile /content/pages/Search_Resume.py

import streamlit as st
import plotly.graph_objects as go
from spacy import displacy
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm
import streamlit as st
import plotly.graph_objects as go
import json
from flask import Flask, request, jsonify
from deta import Deta
import time
from PIL import Image
import textwrap
from fpdf import FPDF
from google.colab import files


#Page title 
page_title = "Document Management System"
page_icon = ":open_file_folder: " # emojis: https://www.webfx.com/tools/emoji-cheat-sheet/
layout = "wide"


st.set_page_config(page_title = page_title, page_icon=page_icon, layout=layout)
st.title(page_title + " " + page_icon)

image = Image.open('/content/legend.png')

guide = st.checkbox('Colour Legend')

if guide:
    st.image(image, width=250)

count=0 #for button widget 


#==========================================================================================================================
#==========================================================================================================================


#Cache model to avoid reloading every run
@st.cache(allow_output_mutation=True) 



# --->Load CUSTOM NER Model
def load_model():
  nlp = spacy.load("/content/drive/MyDrive/NLP_Asg/output/output60/content/output/model-best")
  return(nlp)

nlp = load_model()



# --->Load SPACY NER Model
def load_spacy_model():
  spacy_nlp = spacy.load("en_core_web_sm")
  return(spacy_nlp)

spacy_nlp = load_spacy_model()










### --->Update then Opening JSON file containing Entity and Values
def load_data(filename): 
    entities_dict_list = json.load(open(filename))
    return entities_dict_list

ent_val_dict_json = load_data('/content/ent_val_dict')






### --->DATABASE

def load_database(DATA_KEY):
  deta = Deta(DATA_KEY)
  db = deta.Base('DMS_db')
  return db 

db = load_database("c0l66t56_puXQuHpDcqLmB43nBU69orZDc9wgcLbf")


### --->DOWNLOAD RESUME


def text_to_pdf(text, filename):
    a4_width_mm = 210
    pt_to_mm = 0.35
    fontsize_pt = 10
    fontsize_mm = fontsize_pt * pt_to_mm
    margin_bottom_mm = 10
    character_width_mm = 7 * pt_to_mm
    width_text = a4_width_mm / character_width_mm

    pdf = FPDF(orientation='P', unit='mm', format='A4')
    pdf.set_auto_page_break(True, margin=margin_bottom_mm)
    pdf.add_page()
    pdf.set_font(family='Courier', size=fontsize_pt)
    splitted = text.split('\n')

    for line in splitted:
        lines = textwrap.wrap(line, width_text)

        if len(lines) == 0:
            pdf.ln()

        for wrap in lines:
            pdf.cell(0, fontsize_mm, wrap, ln=1)

    pdf.output(filename, 'F')




#def download_pdf():
 # files.download('output_pdf_text.pdf')
#download_pdf()


def remove_non_ascii_1(text):
  return ''.join([i if ord(i) < 128 else ' ' for i in text])

#==========================================================================================================================
#==========================================================================================================================


#Search field  
search_item = st.text_input(label='Search Entity')      #PREPROCESS LATERRRRRRRRRRRRRRRR 
search_item_ori = search_item
if search_item is None:
  st.error('No search entry found.')

elif search_item:
  with st.spinner('Wait for it...'):
    time.sleep(5)

  search_ent = ""



  #Extract entity of search_item IN DATABASE
  search_item = search_item.lower()
  

  for key, value in ent_val_dict_json.items():  #extract entity of search_item from json
    for val in value:                           #deep search because dictionary might have nested lists 
        if val==search_item:
          search_ent=key
          break
        for v in val:
          if v==search_item:
            search_ent=key
            break
          for v2 in v:
            if v2==search_item:
              search_ent=key
              break
            for v3 in v2:
              if v3==search_item:
                search_ent=key
                break


  
  if search_ent=="":     #if entity not found in database, find generally with CUSTOM MODEL         
    
    #Extract entity of search_item
    similar_doc = nlp(search_item_ori) #keep original letter cases
    for ent in similar_doc.ents: #assuming that user passes 1 Noun 
      search_ent = str(ent.label_)

    st.text("Custom model search: " + search_ent)

    if search_ent=="":  #find Generally with SPACY MODEL

      similar_doc = spacy_nlp(search_item_ori) #keep original letter cases
      for ent in similar_doc.ents: #assuming that user passes 1 Noun 
        search_ent_spacy = str(ent.label_)

      if search_ent_spacy == "GPE":
        search_ent = "Location"
      elif search_ent_spacy == "PERSON":
        search_ent = "Name"
      elif search_ent_spacy == "ORG":
        search_ent = "Companies worked at"
    
      st.text("Spacy model search: " + search_ent)


    if search_ent =="":
        st.error('Results not found.', icon="🚨")
 
    else:
        st.subheader("Showing similar results...") #other record in database with same search_ent 
        fetch_str = "entities." + search_ent + "?not_contains" #other than search item of 2015


        #st.text(fetch_str + ":" + search_item_ori)

        def fetch_items():
            res = db.fetch({fetch_str:search_item_ori},limit=3) #return 3 similar results 
            return res.items

        if len(fetch_items())==0:
          st.error('Results not found.', icon="🚨")

        #find documents with search_item belonging to its entity 
        for i in range(len(fetch_items())):  
          render_text = fetch_items()[i]['text']
          doc =  nlp(render_text)

          colors = {search_ent: "linear-gradient(90deg, yellow, red)"}
          options = {"ents": [search_ent], "colors":colors}
          ent_html = displacy.render(doc, style="ent", jupyter=False, options=options) #return document with highlighted entity
          # Display the entity visualization in the browser:
          st.markdown(ent_html, unsafe_allow_html=True)
          st.text(" ")


          ######### DOWNLOAD AND VIEW RESUME 

          
          count += 1
          if st.button("View Resumé", key=count):
            
            sample_text = render_text 

            sample_text = remove_non_ascii_1(sample_text)

            #write to file
            text_file = open('input.txt', "w")
            text_file.write(sample_text)
            text_file.close()

            input_filename = 'input.txt'
            output_filename = 'output_pdf_text.pdf'
            file = open(input_filename)
            text = file.read()
            file.close()

            text_to_pdf(text, output_filename)

            with open("output_pdf_text.pdf", "rb") as pdf_file:
              PDFbyte = pdf_file.read()

    
            st.download_button(label="Download",
                      data=PDFbyte,
                      file_name="output_pdf_text.pdf",
                      mime='application/octet-stream')



                
          st.text("================================================================================================================================================================================================")
          st.text(" ")

  else: 
    fetch_str = "entities." + search_ent + "?contains" #eg: entities.DATE?contains:"2007"
    #st.text(fetch_str + ":" + search_item_ori) 

    def fetch_items():
        res = db.fetch({fetch_str:search_item_ori}, limit=3)
        return res.items

    if len(fetch_items())==0:
      st.error('Results not found.', icon="🚨")
     
    #find documents with search_item belonging to its entity 
    for i in range(len(fetch_items())):  
      render_text = fetch_items()[i]['text']
      doc =  nlp(render_text)

      colors = {'Name': "#C8DF52", 
      "Designation": "#9DB6CC", 
      "College Name": "#FFA384" , 
      "Companies worked at" : "#EFE7BC", 
      "Degree" : "#FFC2C7", 
      "Email Address" : "#B6E5D8", 
      "Graduation Year" : "#94C973", 
      "Location": "#F0A160", 
      "Skills": "#D3BBDD", 
      "Years of Experience": "#BBE7FE"}

      options = {"ents": [search_ent], "colors":colors}
      ent_html = displacy.render(doc, style="ent", jupyter=False, options=options) #return document with highlighted entity (exact value) 
      # Display the entity visualization in the browser:
      st.markdown(ent_html, unsafe_allow_html=True)
      st.text(" ")

      ######### DOWNLOAD AND VIEW RESUME 

        
      count += 1
      if st.button("View Resumé", key=count):
          
          sample_text = render_text 

          sample_text = remove_non_ascii_1(sample_text)

          #write to file
          text_file = open('input.txt', "w")
          text_file.write(sample_text)
          text_file.close()

          input_filename = 'input.txt'
          output_filename = 'output_pdf_text.pdf'
          file = open(input_filename)
          text = file.read()
          file.close()

          text_to_pdf(text, output_filename)

          with open("output_pdf_text.pdf", "rb") as pdf_file:
            PDFbyte = pdf_file.read()

  
          st.download_button(label="Download",
                    data=PDFbyte,
                    file_name="output_pdf_text.pdf",
                    mime='application/octet-stream')

      st.text("================================================================================================================================================================================================")
      st.text(" ")
    


```
